## Using Autoencoders

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
```

### Data Prep.

```python
# Assume ratings_df is your ratings dataframe
user_item_matrix = ratings.pivot(index='userId', columns='movieId', values='rating')
user_item_matrix.fillna(0, inplace=True)

# Normalize the data by subtracting mean
user_ratings_mean = np.mean(user_item_matrix.values, axis=1)
user_item_matrix_demeaned = user_item_matrix.values - user_ratings_mean.reshape(-1, 1)

X_train, X_test = train_test_split(user_item_matrix_demeaned, test_size=0.2, random_state=42)
```

### Build the autoencoder

```python
# Model parameters
input_dim = X_train.shape[1]  # number of movies
hidden_dim = 512
encoded_dim = 256

# Input
input_layer = Input(shape=(input_dim,))

# Encoder
encoder = Dense(hidden_dim, activation='relu', activity_regularizer=tf.keras.regularizers.l2(1e-5))(input_layer)
encoder = Dropout(0.5)(encoder)
encoder = Dense(encoded_dim, activation='relu')(encoder)

# Decoder
decoder = Dense(hidden_dim, activation='relu')(encoder)
decoder = Dropout(0.5)(decoder)
decoder = Dense(input_dim, activation='linear')(decoder)  # linear activation for reconstruction

# Compile model
autoencoder = Model(inputs=input_layer, outputs=decoder)
autoencoder.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')
```

### Train the autoencoder

```python
history = autoencoder.fit(
    x=X_train,
    y=X_train,
    epochs=50,
    batch_size=256,
    shuffle=True,
    validation_data=(X_test, X_test)
)
```

```python
user_item_matrix_pred = autoencoder.predict(user_item_matrix_demeaned)
recommendations = user_item_matrix_pred + user_ratings_mean.reshape(-1, 1)
```

```python
def get_top_n_recommendations(user_id, n=10):
    recommended_items = pd.DataFrame(recommendations[user_id], index=user_item_matrix.columns, columns=['rating'])
    top_n = recommended_items.sort_values(by='rating', ascending=False).head(n)
    return top_n.index.tolist()
```
